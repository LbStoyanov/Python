```python
import seaborn as sns
```

```python
#allow to see vizualizations inside of the notebook
%matplotlib inline 
```

```python
tips = sns.load_dataset('tips') #build-in in seaborn library
tips.head()
```

```python
#distplot type => allows to show the distribution f a univariate set of observetions or in other words one variable => DEPRICATED!
sns.histplot(tips)
```

```python
#jointpots
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
#pairplot
sns.pairplot(tips,hue='sex', palette='coolwarm')
```

```python
#rugplots
sns.rugplot(tips['total_bill'])
```
