```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

## Titanic Data Visualization Task

Recreate the plots below using the **Titanic** dataframe.

There are very few hints, since most of the plots can be created with just one or two lines of code â€” a hint would basically give away the solution.

Pay close attention to the **x** and **y** labels in each plot, as they may offer subtle hints.


```python
sns.jointplot(x='fare', y='age',data=titanic)
```

```python
sns.displot(titanic['fare'],kde=False,color='red',bins=20)
```

```python
class_colors = {
    'First': 'blue',
    'Second': 'lightgreen',  
    'Third': 'red'    
    
}

sns.boxplot(x='class',y='age', data=titanic, palette='rainbow') # We can use the class_color for custom colors

```

```python
sns.swarmplot(x='class',y='age', data=titanic,palette='Set2')
```

```python
colors = {
    'male': 'lightblue',
    'female': 'lightgreen', 
}

sns.countplot(x='sex',data=titanic,palette=colors)
```

```python
# Convert boolean columns to integers (True -> 1, False -> 0)
titanic['adult_male'] = titanic['adult_male'].astype(int)
titanic['alone'] = titanic['alone'].astype(int)

# Select all numeric columns (now including converted boolean columns)
numeric_data = titanic.select_dtypes(include='number')

# Compute correlation matrix
corr_matrix = numeric_data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Titanic Numeric & Boolean Features')
plt.show()
```
